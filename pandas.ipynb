```python
# How many male and female passenger were onboarded
m = df['Sex'].loc[df['Sex'] == 'male'].count()
f = df['Sex'].loc[df['Sex'] == 'female'].count()
print("Male:", m)
print("Female:", f)
```

```python
df['Sex'].value_counts()
```

```python
#how many survived
df['Survived'].loc[df['Survived'] == True].count()
```

```python
# how many casuality
df['Survived'].loc[df['Survived'] != True].count()
```

```python
# name of the person who is eldest one
df.loc[df['Age'] == max(df['Age'])]['Name']
```

```python
eldest = df.sort_values(by = 'Age').iloc[-1]
eldest[3]
```

```python
# how many passengers in first, second and third class
df.pivot_table(index = 'Pclass', values = 'PassengerId', aggfunc = 'count')
```

```python
# how many people's name starts with 's'
len(df[df['Name'].str[0].isin(['S', 's'])])
```

```python
# try to create a new column which is summation of 'SibSp' and 'parch'
df['Family'] = df['SibSp'] + df['Parch']
df
```

```python
df['SibSp'].dtypes
```

```python
# How many ppl having age less than 25 
df['PassengerId'].loc[df['Age'] < 25].count()
```

```python
# how many ppl dies whose age was less than 40
df['Name'][(df['Age'] < 40) & (df['Survived'] == 0)].count()
```

```python
# from a cabin column separate text & numeric value
df['Letter'] = df['Cabin'].str.extract('([A-Za-z]+)')
df['Digit'] = df['Cabin'].str.extract('([\d+])')
df
```
